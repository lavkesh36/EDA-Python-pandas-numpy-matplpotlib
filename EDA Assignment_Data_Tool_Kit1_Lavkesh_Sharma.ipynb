```python
def uniplot(df,col,title,hue=None):
    
    sns.set_style('darkgrid')
    sns.set_context('poster')
    plt.rcParams['axes.titlesize'] = 40
    plt.rcParams['axes.titlepad'] = 28
    plt.rcParams['axes.labelsize'] = 22
    
    temp = pd.Series(data=hue)
    fig,ax = plt.subplots()
    width = len(df[col].unique()) + 8 + 4.5*len(temp.unique())
    fig.set_size_inches(width ,8)
    plt.xticks(rotation=50)
    plt.yscale('log')
    plt.title(title)
    ax = sns.histplot(data=df , x=col ,hue=hue, palette='magma')
    plt.show()
    
```

```python
target_0.head()
```

```python
def plotuninewapp(var):

    plt.style.use('ggplot')
    sns.despine
    fig,(ax1,ax2) = plt.subplots(1,2,figsize=(20,6))
    
    sns.countplot(x=var, data=target_0,ax=ax1)
    ax1.set_ylabel('Total Counts')
    ax1.set_title(f'Distribution of {var} for Non-Defaulters',fontsize=20)
    ax1.set_xticklabels(ax1.get_xticklabels(), rotation=20, ha="right")
    
    # Adding the normalized percentage for easier comparision between defaulter and non-defaulter
    for p in ax1.patches:
        ax1.annotate('{:.1f}%'.format((p.get_height()/len(target_0))*100), (p.get_x()+0.25, p.get_height()+100))
        
    sns.countplot(x=var, data=target_1,ax=ax2)
    ax2.set_ylabel('Total Counts')
    ax2.set_title(f'Distribution of {var} for Defaulters',fontsize=25)    
    ax2.set_xticklabels(ax2.get_xticklabels(), rotation=20, ha="right")
    
    # Adding the normalized percentage for easier comparision between defaulter and non-defaulter
    for p in ax2.patches:
        ax2.annotate('{:.1f}%'.format((p.get_height()/len(target_1))*100), (p.get_x()+0.1, p.get_height()+50))
    
    plt.show()
```
